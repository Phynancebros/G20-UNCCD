```python
import rasterio

tif_path = "Datasets_Hackathon/Modis_Land_Cover_Data/2010LCT.tif"  # The TFW is named my_raster.tfw
with rasterio.open(tif_path) as src:
    print(src.crs)       # Might be None if no projection is specified
    print(src.transform) # Should reflect the TFW parameters
    data = src.read(1)   # This is your rasterâ€™s pixel data
```

```python
import matplotlib.pyplot as plt

plt.imshow(data, cmap='viridis')
plt.title("Geo-referenced Raster from TIF + TFW")
plt.colorbar(label='Pixel values')
plt.show()
```
